# Item-based Collaborative Filtering ( 아이템 기반 영화추천)

```python
movie_rate=pd.concat( [neflix_rating,watch_rating ,tiving_rating,disney_ratings])
```

```python
movie_rate=movie_rate.fillna(0)
```

```python
movie_rate.columns.name="TITLE"
```

```python
movie_rate=movie_rate.set_index("nickname")
```

```python
movie_rate
```

```python
movie_user_rate = movie_rate.values.T
movie_user_rate.shape
```

```python
SVD = TruncatedSVD(n_components=12, random_state=3)
matrix = SVD.fit_transform(movie_user_rate)
matrix.shape
```

```python
matrix
```

```python
corr = np.corrcoef(matrix)
corr[np.isnan(corr)]  = 0
corr
```

```python
movie_title = movie_rate.columns
movie_title_list=list(movie_rate.columns)
```

```python
def get_recommend_movie_by_item(title) : 
    coffey_hands = movie_title_list.index(title)
    corr_coffey_hands=corr[coffey_hands]
    recomend_movie_list=list(movie_title[(corr_coffey_hands<1.0) & (corr_coffey_hands>0.9)])
    dict_movie ={}
    for x in recomend_movie_list :
        dict_movie[x] = movie_contents[movie_contents["TITLE"] == x].values[0][3]
    return dict_movie
```

### 코사인유사도를 이용하여 영화 제목의 유사한 장르를 가진 영화 추천

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def convert(x):
    i=''
    if  not isinstance(x,float) : 
        i = x.split(',')
    return ' '.join(i)
```

```python
movie_contents['GENRES'] = movie_contents['GENRES'].apply(convert)
```

```python
movie_contents
```

```python
transformer = CountVectorizer()
genres_vector = transformer.fit_transform(movie_contents['GENRES'])
genres_vector.shape
```

```python
similarity = cosine_similarity(genres_vector, genres_vector)
similarity
```

```python
similarity = similarity.argsort()
similarity
```

```python
similarity = similarity[:, ::-1]
similarity
```

```python
def get_recommend_movie_list_genre( movie_genre,top=50):
    
    search_df = movie_contents[movie_contents['GENRES'] == movie_genre ]
    search_df_index = search_df.index.values
    similarity_index = similarity[search_df_index, :top].reshape(-1)
    similarity_index = similarity_index[similarity_index != search_df_index] #제목이 movie_title 인 영화 제외
   
    result = movie_contents.iloc [similarity_index]
    
    result = result[:2]
    movie_dict ={}
    for idx, row in result.iterrows():        
        movie_dict[row["TITLE"]] =  row["OTT"]
    
    
    return movie_dict
```

```python
def get_recommend_movie_list_title( genre, ott,top=50):
    
    search_df = movie_contents[movie_contents['TITLE'] == genre]
    search_df_index = search_df.index.values
    similarity_index = similarity[search_df_index, :top].reshape(-1)
    similarity_index = similarity_index[similarity_index != search_df_index] #제목이 movie_title 인 영화 제외
   
    result = movie_contents.iloc [similarity_index]

    if  ott in list( result["OTT"]) :
        result = result[result["OTT"] == ott ] 
    
    result = result[:2]
    movie_dict ={}
    for idx, row in result.iterrows():        
        movie_dict[row["TITLE"]] =  row["OTT"]
    
    
    return movie_dict
```

## 사용자 응답에 대한 결과 확인

```python
ott_suvery =pd.read_excel(r"D:\Python\TM\TeamProject/ott_survey.xlsx")
```

```python
ott_suvery=ott_suvery.drop(['Unnamed: 0','index'],axis=1)
```

```python
ott_suvery 
```

```python
def merge_two_dicts(x, y):
 
    z = x.copy()
    
    z.update(y)
    
    return z
```

```python
user_movie=[]
for idx, row in ott_suvery.iterrows():
    pre_title  = row["선호콘텐츠"].split(',')
    
    pre_dict_all ={}
    for title in pre_title :
        ott_list = row["OTT_LIST"].replace('[','').replace(']','').split(',')
        
        if title != "없음" :
            pre_dict = get_recommend_movie_by_item(title.strip())
           
            for ott in ott_list :
                pre_dict1 = get_recommend_movie_list_title(title,ott)
                if(bool(pre_dict1)) :
                    pre_dict = merge_two_dicts(pre_dict,pre_dict1)
            pre_dict_all = merge_two_dicts(pre_dict_all,pre_dict )
                
    user_movie.append(pre_dict_all)
    
    
```

```python
user_movie
```

```python
#OTT 플랫폼에 해당하는 대표 프로그램 정보 가져오기
def getOTTMainProgram(ott_list) :
    ott_preference = pd.read_excel(r"D:\Python\TM\TeamProject/OTT프로그램정리.xlsx")
    
    main_programs = {}

    for i in ott_list :
        
        ott_movies = ott_preference[ott_preference["OTT프로그램"]== i.strip()].values[0][4]
        for x in ott_movies.split(','):
            main_programs[x.strip()] = i.strip()
            
    return main_programs
```

```python
for idx, row in ott_suvery.iterrows(): 
    ## 선호콘텐츠가 없는 경우 OTT_LIST의 대표 프로그램을 세팅
    if bool(user_movie[idx]) ==False :

        ott = row["OTT_LIST"]
        ott = ott.replace('\'','')
        ott = ott.replace('[','')
        ott = ott.replace(']','')
        ott_list = ott.split(',')
        
        dict_movie=getOTTMainProgram(ott_list)
        user_movie[idx].update(dict_movie)
```

```python
user_recomend={}
for idx, row in ott_suvery.iterrows(): 
    ott = row["OTT_LIST"]
    ott = ott.replace('\'','')
    ott = ott.replace('[','')
    ott = ott.replace(']','')
    ott_list = ott.split(',')
    user_recomend[row["USER_ID"]] = [ ott_list ,user_movie[idx]]
```
